```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

ecom = pd.read_csv("../data/processed/ecom_full_cleaned.csv")
bank = pd.read_csv("../data/processed/bank_full_cleaned.csv")
# univariate analysis
# E-Commerce: Fraud distribution
sns.countplot(x="class", data=ecom)
plt.title("Fraud Distribution: E-Commerce")
plt.show()

# Bank: Transaction amount distribution
sns.histplot(bank["Amount"], bins=40, kde=True)
plt.title("Bank Transaction Amounts")
plt.show()

# Bivariate analysis
# E-Commerce: Time since signup vs fraud
sns.boxplot(x="class", y="time_since_signup", data=ecom)
plt.title("Signup Delay vs Fraud Class")
plt.show()

```
